```python
def process_simulation(name=model_default):
    print("processing simulation : "+name)
    for day in range(1,6): 
        date = "2018070"+str(day)
        #download the file: 
        command = "scp bijanf@curta.zedat.fu-berlin.de:"+ \
        dir2modelrun+name+prefix+"output/"+ \
        file_prefix+date+"_xb"+file_suffix+" ."
        os.system(command)
        command = "scp bijanf@curta.zedat.fu-berlin.de:"+ \
        dir2modelrun+name+prefix+"output/"+ \
        file_prefix+date+"_xf"+file_suffix+" ."
        os.system(command)
        print("   processing the     " +date)
        cmd = "cdo remapbil,grids LE_v3.0.002_conc-sfc_"+date + \
        "_xb.nc LE_v3.0.002_conc-sfc_"+ date+"_xb_remap.nc"
        os.system(cmd)
        cmd = "cdo remapbil,grids LE_v3.0.002_conc-sfc_"+date + \
        "_xf.nc LE_v3.0.002_conc-sfc_"+date+"_xf_remap.nc"
        os.system(cmd)
        # xb -------------------------------------------------------
        #NO2_URBG :
        cmd  = "cdo mul LE_v3.0.002_conc-sfc_" + \
        date+"_xb_remap.nc  NO2_URBG_mask.nc LE_v3.0.002_conc-sfc_" + \
        date+"_xb_remap_NO2_URBG"+name+".nc"
        os.system(cmd)
        #NO2_SUBG :
        cmd  = "cdo mul LE_v3.0.002_conc-sfc_"+ \
        date+"_xb_remap.nc  NO2_SUBG_mask.nc LE_v3.0.002_conc-sfc_"+ \
        date+"_xb_remap_NO2_SUBG"+name+".nc"
        os.system(cmd)
        #NO2_RUBG :
        cmd  = "cdo mul LE_v3.0.002_conc-sfc_"+ \
        date+"_xb_remap.nc NO2_RUBG_mask.nc LE_v3.0.002_conc-sfc_"+ \
        date+"_xb_remap_NO2_RUBG"+name+".nc"
        os.system(cmd)
         # xf -------------------------------------------------------
        #NO2_URBG :
        cmd  = "cdo mul LE_v3.0.002_conc-sfc_"+ \
        date+"_xf_remap.nc  NO2_URBG_mask.nc LE_v3.0.002_conc-sfc_"+ \
        date+"_xf_remap_NO2_URBG"+name+".nc"
        os.system(cmd)
        #NO2_SUBG :
        cmd  = "cdo mul LE_v3.0.002_conc-sfc_"+ \
        date+"_xf_remap.nc  NO2_SUBG_mask.nc LE_v3.0.002_conc-sfc_"+ \
        date+"_xf_remap_NO2_SUBG"+name+".nc"
        os.system(cmd)
        #NO2_RUBG :
        cmd  = "cdo mul LE_v3.0.002_conc-sfc_"+ \
        date+"_xf_remap.nc NO2_RUBG_mask.nc LE_v3.0.002_conc-sfc_"+ \
        date+"_xf_remap_NO2_RUBG"+name+".nc"
        os.system(cmd)
    # merge all together: 
    #xf
    cmd = "cdo mergetime LE_v3.0.002_conc-sfc_2018070?_xf_remap_NO2_URBG"+name+".nc model_xf_URBG"+name+".nc "
    os.system(cmd)
    cmd = "cdo mergetime LE_v3.0.002_conc-sfc_2018070?_xf_remap_NO2_SUBG"+name+".nc model_xf_SUBG"+name+".nc "
    os.system(cmd)
    cmd = "cdo mergetime LE_v3.0.002_conc-sfc_2018070?_xf_remap_NO2_RUBG"+name+".nc model_xf_RUBG"+name+".nc "
    os.system(cmd)
    #xb
    cmd = "cdo mergetime LE_v3.0.002_conc-sfc_2018070?_xb_remap_NO2_URBG"+name+".nc model_xb_URBG"+name+".nc "
    os.system(cmd)
    cmd = "cdo mergetime LE_v3.0.002_conc-sfc_2018070?_xb_remap_NO2_SUBG"+name+".nc model_xb_SUBG"+name+".nc "
    os.system(cmd)
    cmd = "cdo mergetime LE_v3.0.002_conc-sfc_2018070?_xb_remap_NO2_RUBG"+name+".nc model_xb_RUBG"+name+".nc "
    os.system(cmd)
    # remove the others : 
    cmd = "rm -f LE_v3.0.002_conc-sfc_2018070*_x*_remap_NO2_*BG.nc "
    os.system(cmd)
    cmd = "rm -f LE_v3.0.002_conc-sfc_2018070*_x*_remap.nc "
    os.system(cmd)
    cmd = "rm -f LE_v3.0.002_conc-sfc_2018070*_x*.nc "
    os.system(cmd)
        
```

```python
def plot_scat_model(x,y,color, marker='o', alpha=.5, label="label"):
    
    b, m = polyfit(x, y, 1)
    print(b)
    plt.scatter(x, y, marker=marker,color=color, alpha=alpha, label=label)
    plt.plot(x, b + m * x, '-',color=color)
    
```
